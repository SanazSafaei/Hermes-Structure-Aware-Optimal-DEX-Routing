```python
import pandas as pd
```

```python
df = pd.read_csv("/workspaces/uniswap-v3-crawler/data/aggregate/aggregated_1749964745.csv.gz", compression="gzip")
df.head()
```

```python
df["preprocessing_time"] = pd.to_numeric(df["preprocessing_time"], errors="coerce")
df[["block_number", "algorithm_name", "top_coins", "preprocessing_time", "algorithm_execution_time"]].groupby(["algorithm_name", "top_coins"]).agg({
        "preprocessing_time": ["mean", "min", "max"],
        "algorithm_execution_time": ["mean", "min", "max"]
    }).round(5)
```
