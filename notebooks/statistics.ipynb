```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import os
import subprocess
```

```python
df_stats = pd.read_csv('/workspaces/uniswap-v3-crawler/data/results/stats_20250427_145751.csv')
df_stats = df_stats.sort_values('block_number')
df_stats
```

```python
# Create histograms for num_nodes and treewidth
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot histogram for num_nodes
df_stats['num_nodes'].hist(bins=200, ax=axes[0], color='blue', alpha=0.7)
axes[0].set_title('Histogram of num_nodes (avg: {:.2f})'.format(df_stats['num_nodes'].mean()))

# Plot histogram for treewidth
df_stats['treewidth'].hist(bins=20, ax=axes[1], color='green', alpha=0.7)
axes[1].set_title('Histogram of treewidth (avg: {:.2f})'.format(df_stats['treewidth'].mean()))

# Display the plots
plt.tight_layout()
plt.show()
```
