```python
import sys
import os

# Add the root directory to the Python environment
root_dir = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(root_dir)
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the CSV file
df = pd.read_csv("../data/update_stats_new.csv.gz", compression="gzip")
# df = df[(df['block_number'] >= 22820000) & (df['block_number'] <= 22839999)]
df['update_price'].mean()
df['new_token'].mean()
df['new_pool'].mean()

print(f"update_price: {df['update_price'].describe()}")
print(f"new_token: {df['new_token'].describe()}")
print(f"new_pool: {df['new_pool'].describe()}")
```

```python
# Create histogram plot of update_price
plt.figure(figsize=(16, 8))
plt.xlim(0, df['update_price'].max())
# plt.ylim(0, 10000)
plt.hist(df['update_price'], bins=100, alpha=0.7, edgecolor='black')
plt.xlabel('Number of Updates', fontsize=34)
plt.ylabel('Number of Blocks', fontsize=34)

plt.xticks(fontsize=24)  # Rotate x-axis labels 45 degrees
plt.yticks(fontsize=24)
plt.grid(True, alpha=0.3)

plt.savefig('../data/histogram_update_price.pdf', 
            dpi=500, 
            format='pdf', 
            bbox_inches='tight',
            facecolor='white',
            edgecolor='none')
plt.show()

# Print some additional statistics
print(f"Total records: {len(df)}")
print(f"Mean update_price: {df['update_price'].mean():.2f}")
print(f"Median update_price: {df['update_price'].median():.2f}")
print(f"Max update_price: {df['update_price'].max()}")
print(f"Min update_price: {df['update_price'].min()}")


```
